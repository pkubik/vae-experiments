```python
from tqdm import tnrange, tqdm_notebook
from time import sleep
import numpy as np
import matplotlib.pyplot as plt
import io
import IPython.core.display as ipyd
```

The code below may be replace with a procedure for training a neural network and continuously updating some important plot, e.g. reconstruction of the input image in VAE or GANs.

```python
plt.interactive(False)

def create_image():
    with io.BytesIO() as buff:  # this seems overly complicated but I didn't find a simpler way of plotting to buffer
        plt.imshow(np.random.randint(0, 100, [32, 32, 3]).astype(np.float32))  # just random 32x32 RGB image
        plt.savefig(buff, format='svg')
        buff.seek(0)
        image_data = buff.read().decode()
        plt.close()  # prevent Jupyter from eagerly displaying the plot
    return image_data

svg = ipyd.SVG(data=create_image())
svg_display = ipyd.display(svg, display_id=True)  # display the widget, remember the handle for updates

pbar = tnrange(100, desc='training')  # tqdm for the progress bar, might be replaced with direct `tqdm_notebook`
for i in pbar:
    pbar.set_postfix(loss=0.1 / (i+1.), refresh=False)  # update the progress bar, throw in some additional metrics

    svg.data = create_image()  # create updated plot
    ipyd.update_display(svg, display_id=svg_display.display_id)  # update the plot
    
    print("some log")  # some logs to show that there are not affected in any way
    sleep(0.2)
```
